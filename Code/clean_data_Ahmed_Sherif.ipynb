## Data Quality Issues

1. The dataset contains **1,334 duplicate records** that need to be identified and handled.

2. The **PatientId** column is stored as a `float`, while it should be an **integer** since it represents an identifier rather than a numerical value.

3. The **ScheduledDay** and **AppointmentDay** columns are currently stored as **strings**. They should be converted to the **datetime** data type for proper time-based analysis.

4. The **Gender** and **No-show** columns contain **categorical string values with only two categories**. For easier analysis and modeling, they should be **encoded into numerical values** (e.g., `No = 0`, `Yes = 1`).

5. The **Handcap** column contains **five different values (0–4)** representing levels of disability. It may be simplified into a **binary feature**:
   - `1` → Has a disability  
   - `0` → No disability

6. The **AppointmentDay** column **does not include hour information**, as all time values are set to `00:00:00`, which limits time-of-day analysis.

7. The **Age** column contains **negative values**, which are invalid and should be investigated and corrected or removed.

### Solve Issues

```python
med = pd.read_csv('KaggleV2-May-2016.csv')
```

```python
med.columns
```

### Drop Duplicates

```python
med  = med.drop_duplicates(subset=['PatientId', "ScheduledDay", "AppointmentDay","Age","Neighbourhood"], keep='first')
```

```python
med.duplicated(['PatientId', "ScheduledDay", "AppointmentDay","Age","Neighbourhood"],keep= False).sum()
```

### Change Data Types

```python
med['PatientId'] = med['PatientId'].astype(int)
```

```python
med['PatientId'].info()
```

```python
med['AppointmentDay'] = pd.to_datetime(med['AppointmentDay']).dt.date
med['ScheduledTime']= pd.to_datetime(med['ScheduledDay']).dt.time
med['ScheduledDay'] = pd.to_datetime(med['ScheduledDay']).dt.date
```

```python
col = med.pop("ScheduledTime")
med.insert(4, "ScheduledTime", col)
```

```python
med["AppointmentDay"].info()
```

```python
med.head()
```

### Check the appointments

```python
med[med['AppointmentDay'] < med['ScheduledDay']]
```

```python
med = med[med['AppointmentDay'] >= med['ScheduledDay']]
```

```python
med[med['AppointmentDay'] < med['ScheduledDay']]
```

```python
med['AppointmentDay'] = med['AppointmentDay'].astype('date32[pyarrow]')
med['ScheduledDay'] = med['ScheduledDay'].astype('date32[pyarrow]')
```

```python
med["Handcap"]= med["Handcap"].apply(lambda x: 1 if x > 0 else 0)
```

### Remove the negative ages

```python
med=med[med['Age']>=0]
```

```python
med["Age"].value_counts()
```

```python
med[med["Age"] >100]
```

```python
def whisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lower=q1-1.5*iqr
    upper=q3+1.5*iqr
    return lower, upper

lower, upper = whisker(med["Age"])
med["Age"]=np.where(med["Age"] > upper, upper, med["Age"]).astype(int)
```

```python
sms_counts = med.groupby(['SMS_received','No-show']).size().reset_index(name='Count')

fig = px.bar(sms_counts, 
             x='SMS_received', 
             y='Count', 
             color='No-show', 
             barmode='group',
             title='Effect of SMS Reminder on No-show (Count)',
             text='Count',
             labels={'SMS_received':'SMS Received', 'Count':'Number of Appointments'},color_discrete_sequence=["#327B64","#B14A4A"])

fig.show()
```

##  SMS Reminders

* Patients **who received an SMS** have a **27% non-attendance rate**.
* Patients **who did not receive an SMS** have a **16% non-attendance rate**.
* This is **counterintuitive**, as reminders are expected to increase attendance, suggesting that **other factors may influence patient behavior** ,or the way of reminding should be changed.

```python
wait_counts = med.groupby(['WaitingDays','No-show']).size().reset_index(name='Count')

wait_counts = wait_counts[wait_counts['WaitingDays'] <= 60]

fig = px.bar(wait_counts, 
             x='WaitingDays', 
             y='Count', 
             color='No-show', 
             barmode='group',
             title='Number of Appointments by Waiting Days and No-show',
             text='Count',
             labels={'WaitingDays':'Waiting Days', 'Count':'Number of Appointments'},
             color_discrete_sequence=["#327B64","#B14A4A"])

fig.show()
```

##  Waiting Days Analysis

The figure shows that **shorter waiting periods lead to higher attendance rates**.

* Patients with **fewer waiting days** are more likely to attend their appointments.
* As the **waiting time increases**, the likelihood of non-attendance grows, indicating that **long delays negatively impact compliance**.

## Recommendations to Improve Patient Attendance

### 1️ Reduce Waiting Times

* Minimize the gap between appointment booking and the actual visit.
* Shorter waiting periods significantly increase patient attendance.

### 2️ Improve Reminder Systems

* Send timely reminders via SMS, email, or phone calls before appointments.
* Evaluate the most effective communication channels for different patient groups.

### 3️ Focus on Age-Specific Strategies

* Children generally attend with their families, while adolescents and older adults show lower attendance.
* Consider educational outreach for parents of adolescents and additional support for older patients.

### 4️ Support Vulnerable Groups

* Provide assistance such as transportation guidance or scheduling flexibility for patients with chronic conditions or mobility challenges.

### 5️ Monitor and Adjust Scheduling

* Regularly review attendance trends using dashboards.
* Adjust appointment schedules, reminder timing, and patient communication based on observed patterns.

### 6️ Simplify Appointment Access

* Make appointment booking easy through online platforms or phone systems.
* Reduce complexity to prevent missed appointments due to confusion or forgetfulness.

### 7️ Educate Patients

* Share clear messages about the importance of attending appointments and potential health consequences of missing them.
